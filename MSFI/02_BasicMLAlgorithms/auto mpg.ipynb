```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\courses\\data\\auto-mpg-data\\auto-mpg.data.txt",delim_whitespace=True,na_values=['?'],
                names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name'])

df.dtypes
df.columns
```

```python
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)

X = df.loc[:,'cylinders':'origin']
X_1d = pca.fit_transform(X)
X_1d
```

```python
fig,ax = plt.subplots(figsize=(10,8))
ax.scatter(X_1d,df.mpg)
ax.scatter(df.acceleration,df.mpg)
ax.scatter(df.cylinders,df.mpg)
ax.set(xlabel='1D proj of cylinders to origin', ylabel='mpg')
```

```python
%matplotlib notebook

from mpl.toolkits.matplot3d import Axes3D
```

```python
from sklearn.model_selection import train_test_split

y = df.mpg
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=.1, random_state=42)
#X is a matrix and y is a scalar

print(train_X.shape, test_X.shape, train_y.shape, test_y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler()
y_scaler = StandardScaler()

X_scaler.fit(train_X)
train_X_scaled = X_scaler.transform(train_X)
test_X_scaled = X_scaler.transform(test_X)


y_temp = train_y.values.reshape(-1,1)
y_scaler.fit(y_temp)
train_y_scaled = y_scaler.transform(y_temp)
test_y_scaled = y_scaler.transform(test_y.values.reshape(-1,1))

type(y_scaler)
```

```python
train_X.describe()
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score

lin = LinearRegression()
lin.fit(train_X_scaled, train_y_scaled)

pred_scaled = lin.predict(test_X_scaled)
print(mean_squared_error(test_y_scaled, pred_scaled))
print(r2_score(test_y_scaled, pred_scaled))

lin = LinearRegression()
lin.fit(train_X, train_y)

pred = lin.predict(test_X)
print(mean_squared_error(test_y, pred))
print(r2_score(test_y, pred))

print(lin.coef_,lin.intercept_)
```

```python
sgd = SGDRegressor(verbose=True, max_iter=1000, tol=1e-4, verbose=True)
sgd.fit(train_X_scaled, train_y_scaled.ravel())

pred_scaled_sgd = sgd.predict(test_X_scaled)
print(mean_squared_error(test_y_scaled, pred_scaled_sgd))
print(r2_score(test_y_scaled, pred_scaled))

```
